```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("training_twitter.csv")
data1=pd.read_csv("test_twitter.csv")
```

```python
data.head()
```

```python
#training_data
x_train=data['text']
y_train=data['airline_sentiment']
```

```python
#testing data
x_test=data1['text']
```

```python
#splitting the data
def split(data):
    document=[(data[i].split())for i in range(len(data))]
    return document
```

```python
#splitting the testing data
x_train_data=split(x_train)
len(x_train_data)
```

```python
#splitting the testing data
x_test_data=split(x_test)
len(x_test_data)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stop=stopwords.words('english')
stop+=list(string.punctuation)
stop
```

```python
def clean_data(words):
    output_words=[]
    for w in words:
        if w.lower() not in stop and not w.startswith('@'):
            pos=pos_tag(w)
            clean_words=lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
    return output_words
```

```python
#fetching the clear training data
x_train_documents=[(clean_data(document))for document in x_train_data]
```

```python
len(x_train_documents)
```

```python
#fetching the clean testing data
document_x_test=[clean_data(doc)for doc in x_test_data]
```

```python
document_x_test[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#joining the training data to form groups
train_documents=[" ".join(doc)for doc in x_train_documents]
```

```python
#joining the testing data to form groups
test_documents=[" ".join(doc)for doc in document_x_test]
test_documents
```

```python
count_vec=CountVectorizer(max_features=2000)
x_train_features=count_vec.fit_transform(train_documents)
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features=count_vec.transform(test_documents)
```

# Using SVM

```python
svm_clf=svm.SVC()
svm_clf.fit(x_train_features,y_train)
```

```python
y_pred_svm=svm_clf.predict(x_test_features)
```
